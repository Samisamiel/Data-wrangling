```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the raw data and calling the heading
cus_data = pd.read_csv('/content/Customers (2).csv')
cus_data.head()
```

# **Brief Data cleaning**  

```python
#calling the columns to view the one to drop
cus_data.columns
```

```python
#Applying the drop method to drop some columns
q = ['Unnamed: 4','Unnamed: 5', 'Unnamed: 6','Unnamed: 8','Unnamed: 18','Unnamed: 17','Prefix','BirthDate.1']
cus_data.drop(q, axis=1, inplace=True)
```

```python
#checking if the code work
cus_data
```

```python
#Replacing the maritalStatus and Gender column with the right words
cus_data.MaritalStatus = cus_data.MaritalStatus.replace('M', 'Married')
cus_data.MaritalStatus = cus_data.MaritalStatus.replace('S', 'Single')
cus_data.Gender = cus_data.Gender.replace('M', 'Male')
cus_data.Gender = cus_data.Gender.replace('F', 'female')
cus_data.HomeOwner = cus_data.HomeOwner.replace('Y', 'Yes')
cus_data.HomeOwner = cus_data.HomeOwner.replace('N', 'No')
```

```python
#checking if it works
cus_data.head()
```

```python
#Checking the info to know the data type of each column
cus_data.isnull().sum()
```

```python
#Handling null entry
cus_data.fillna(method='ffill', inplace=True)
```

```python
#checking if the code works
cus_data.isnull().sum()
```
